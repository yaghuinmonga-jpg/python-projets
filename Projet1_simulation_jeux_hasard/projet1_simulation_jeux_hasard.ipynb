# Projet 1 - Simulation de jeux de hasard en Python

## Objectif du projet 

L'objectif de ce projet est de simuler des jeux de hasard à l'aide du language Python afin d'analyser leur rentabilité sur un grand nombre d'essaies.

Ce projet vise à :
- Mettre en pratique les bases de Python (variables, boucles, conditions)
- Comprendre la notion d'espérance mathématique
- Analyser si un jeu est gagnant ou perdant sur le long terme

## Contexte

Un joueur dispose d'un capital initial et participe à des jeux de hasard.
Chaque jeu est répété un grand nombre de fois afin d'observer l'évolution du capital.

```python
import random

# Paramètres du jeu
capital_initial = 1000
capital = capital_initial
mise = 10
nb_parties = 100000
historique_capital = []
```

```python
# Simulation du jeu 

for i in range (nb_parties):
    resultat = random.choice([mise, -mise]) # gain ou perte
    capital += resultat
        
```

```python
# Simulation du jeu 

for i in range (nb_parties):
    resultat = random.choice([mise, -mise]) # gain ou perte
    capital += resultat
    historique_capital.append(capital)
```

### Evolution du capital au cours du temps

Le graphique ci-dessous représente l'évolution du capital du joueur au fil des parties.
On y observe une forte volatilité, avec des phases successives de gains et de pertes et ceux malgré une espérance proche de zéro à chaque partie.


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(historique_capital)
```

```python
capital

gain_moyen = (capital - capital_initial) / nb_parties

gain_moyen
```

```python
import numpy as np

variance_capital = np.var(historique_capital)
ecart_type_capital = np.std(historique_capital)

variance_capital, ecart_type_capital
```

### Analyse statistique : variance et risque 

L'ésperance de gain par partie est proche de zéro, ce qui signifie ue le jeu est théoriquement équitable.
Cependant, la varance élevéé du capital montre que les trajectoires possibles sont ttrès dispersées.

L'écart-type important indique une forte volatilitée: le capital peut fortement évoluer à la hausse ou la baisse au cours du temps, malgré une espérance nulle. Cela met en évidence un risque importrant pour le joueur.

Meme dans un jeu à chances égales, le risque de perte importantes existe à cause de l'accumulation des fluctuations aléatoires sur un grand nombre de parties.

```python
import matplotlib.pyplot as plt 
 
plt.figure(figsize=(8, 5))
plt.hist(historique_capital, bins=50)
plt.xlabel("Capital")
plt.ylabel("Fréquence")
plt.title("Distribution du capital au cours de la simulation")
plt.show()
```

### Observation de la distribution du capital

L'histogramme montre que le capital du joueur peut prendre des valeurs très variées durant la simulation, ce qui ce traduit par une forte dispersion des résultats.
On observe que la plupart des valeurs se trouvent dans la zone centrale de l'histogramme et que les valeurs élevées à une plus faible fréquence.
Cela signifie que, meme si des gains importants sont possibles, ils restent statistiquement rares. Et que le capital passe beaucoup plus de temps autours des valeurs intermediaires ni très elevées ni trop faibles.

Cette observation met en évidence le role de la variance: meme dans un jeu à ésperance nulle, les trajectoires du capital peuvent etre très différentes et comporter un risque important pour le joueur.


### Notion de risque de ruine 

Meme si le jeu est théoriquement équitable (ésperance de gain proche de zéro), la simulation met évidence un risque de ruine du joueur.
La forte variance implique que le capital peut subir de longues séries de pertes consécutives, entrainant une baisse importante du capital.

Sur un jeu suffisamment long, il est possible que le capital atteigne un niveau très faible voir proche de zéro. Ceci illustre le fait que l'ésperance seul ne permet pas à lui seul d'évaluer le risque. La volatilité et la gestion du capital jouent donc un role central.
